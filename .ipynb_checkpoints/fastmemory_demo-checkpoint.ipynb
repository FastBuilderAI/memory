# 🧠 FastMemory: The RAG-Replacement for Deterministic AI Agents

Welcome to the **FastMemory** demonstration. This notebook showcases how **Topology Architecture** (Component, Block, Function, Data, Access, Event) replaces traditional Vector RAG to eliminate hallucinations and provide AI agents with a deterministic logical map of enterprise data.

### 🏆 SOTA Performance
FastMemory holds **SOTA status on 11 distinct benchmarks** (FinanceBench, LexGLUE, GraphRAG-Bench, etc.), outperforming standard RAG in massive multi-hop reasoning.

## 1. Setup & Installation

FastMemory is powered by a high-performance **Rust** engine exposed via a Python native module. Install it via `pip`.

```python
# !pip install fastmemory
import fastmemory
import json
print("FastMemory Engine Loaded Successfully!")
```

## 2. Defining the "Topology" (Action-Topology Format)

Unlike RAG, which treats documents as flat strings, FastMemory treats them as **Atomic Text Functions (ATFs)**. Below is a sample from an **Enterprise Banking** system.

```python
atf_docs = """
## [ID: auth_core]
**Action:** Validate_Session_JWT
**Logic:** If token is expired, return 401; otherwise, fetch User_Scopes from Redis.
**Data_Connections:** cache:session_store, db:user_roles
**Access:** Role_Authenticated_User
**Events:** On_Auth_Success, On_Token_Refresh

## [ID: loan_processor]
**Action:** Approve_Mortgage_Logic
**Logic:** Calculate Debt-to-Income (DTI). If DTI < 43% and Credit_Score > 640, permit Pre-Approval.
**Data_Connections:** bureau:credit_report, internal:account_history
**Access:** Role_Loan_Officer
**Events:** Lead_Scoring_Complete, Underwriting_Queue_Entry
"""

print("Sample Topology Defined!")
```

## 3. The Clustering Engine (Louvain Community Detection)

Now we pass the raw Topology text into the Rust engine. FastMemory will automatically cluster these principles into **Cognitive Blocks** based on their topological relationships.

```python
# Run the instantaneous clustering engine
topology_json = fastmemory.process_markdown(atf_docs)
memory_graph = json.loads(topology_json)

print(f"Successfully built a memory graph with {len(memory_graph)} clusters!")
```

## 4. Grounding an AI Agent (Deterministic Context)

Standard RAG might retrieve a "random snippet" about loans. FastMemory returns the **exact logical block**. 

When an AI agent needs to answer a question, you provide the relevant **Block** logic as the ground truth.

```python
# Simulating the search for the 'loan_processor' block
for block in memory_graph:
    if 'loan' in block.get('name', '').lower():
        selected_block = block
        break

print("--- AGENT CONTEXT (GROUND TRUTH) ---")
print(json.dumps(selected_block, indent=2))
```

## 5. Conclusion: Buildings vs. Roads

By using **FastMemory**, your AI agent doesn't just see the "roads" (keywords). It sees the "buildings" (Functional logic). 

1. **Auditability**: You can trace exactly which logical code block generated an AI response.
2. **Safety**: Deterministic **Access** rules prevent agents from hallucinating unauthorized data access.
3. **Speed**: Clustered graphs allow for **0-shot multi-hop reasoning** that Vector DBs cannot achieve.
